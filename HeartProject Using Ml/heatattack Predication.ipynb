```python
#seprating the lables and features from the train data set 
heart_train_data=strat_tarin_set.drop("target",axis=1)
heart_train_lables=strat_tarin_set["target"].copy()
```

```python
# Seprating the test lables and test data in the test data set
heart_test_data=strat_test_set.drop("target",axis=1)
heart_test_lables=strat_test_set['target'].copy()
```

```python
#creating the pipeline for the Any missing values in the data set
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
my_pipeline=Pipeline([('imputer',SimpleImputer(strategy="median")),('std_scaler',StandardScaler()),])
```

```python
# Making the model 
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(heart_train_data,heart_train_lables)
clf.predict(heart_test_data)
```

```python
#Finding the accuracy of thr by fnding the mean
from sklearn.model_selection import cross_val_score 
a=cross_val_score(clf,heart_train_data,heart_train_lables,cv=3,scoring="accuracy")
```

```python
a.mean()
```

```python
#EvALATION OF THE CLASSIYFIRE 
#1. WE HAVE TO FIND THE PRIDCATED OUTPUT AND THEN IT WILL BE MADE THEN CONFUSSION MATRIX FOR THE GIVEN ALGORITHUM
from sklearn.model_selection import cross_val_predict
labels_heart_preicted=cross_val_predict(clf,heart_train_data,heart_train_lables,cv=3)
```

```python
#Making the confusssion matrix 
from sklearn.metrics import confusion_matrix
confussionmatrix=confusion_matrix(heart_train_lables,labels_heart_preicted)
```

```python
import numpy as np
import seaborn as sns
sns.heatmap(confussionmatrix/np.sum(confussionmatrix), annot=True, 
            fmt='.2%', cmap='Blues')
```
