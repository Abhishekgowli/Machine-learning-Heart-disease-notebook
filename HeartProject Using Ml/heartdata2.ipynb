```python
# Spliting the  datafram in train and test set
heatdata_x=heartdata.drop(['target'],axis=1)
```

```python
heartdata_y=heartdata['target'].copy()
```

```python
# Split the data into test and train set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(heatdata_x,heartdata_y,test_size=0.2,random_state=5)
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
#Featature Scaling the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#Bilding the M L Model  
#1. logistic regession
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)
y_predi_lin=clf.predict(x_test)
accuracy_score(y_test,y_predi_lin)
```

```python
# Trainde with Standad scale Data
Clf=LogisticRegression(random_state=51,penalty='l1')
clf.fit(x_train_sc,y_train)
y_pred_lr_sc=clf.predict(x_test_sc)
accuracy_score(y_test,y_pred_lr_sc)
```

```python
# train the data with the support vector machine
from sklearn.svm import SVC
svc_classfier=SVC()
svc_classfier.fit(x_train_sc,y_train)
y_pred_svr_sc=svc_classfier.predict(x_test_sc)
accuracy_score(y_test,y_pred_svr_sc)
```

```python
# taining the KNN Classfire
from sklearn.neighbors import KNeighborsClassifier
knn_classifire=KNeighborsClassifier(n_neighbors=5)
knn_classifire.fit(x_train_sc,y_train)
y_pred_knn_sc=knn_classifire.predict(x_test_sc)
accuracy_score(y_test,y_pred_knn_sc)
```

```python
#random Forest Classfire
from sklearn.ensemble import RandomForestClassifier
rfc_classifire=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=5)
rfc_classifire.fit(x_train_sc,y_train)
y_pred_rfc_sc=rfc_classifire.predict(x_test_sc)
accuracy_score(y_test,y_pred_rfc_sc)
```

```python
#desssion tree calssifire
from sklearn.tree import DecisionTreeClassifier
dtc_classifire=DecisionTreeClassifier(criterion='entropy',random_state=51)
dtc_classifire.fit(x_train_sc,y_train)
y_pred_dtc_sc=dtc_classifire.predict(x_test_sc)
accuracy_score(y_test,y_pred_dtc_sc)
```

```python
cm=confusion_matrix(y_test,y_pred_rfc_sc)
plt.title("heat map of confusion matrix",fontsize=51)
sns.heatmap(cm,annot=True)
plt.show()
```
